# Martingais e Concentração — Notebook
Demonstrações experimentais da desigualdade de Azuma–Hoeffding


```python
import numpy as np
import matplotlib.pyplot as plt

def experiment(n=2000):
    X = np.random.uniform(0,1,n)
    mu = np.mean(X)
    estimates = [np.mean(X[:k]) for k in range(1,n+1)]
    plt.plot(estimates)
    plt.axhline(mu, linestyle='--')
    plt.title('Convergência da média empírica')
    plt.xlabel('n')
    plt.ylabel('Média empírica')
    plt.show()

experiment()
```
